```
#Gather process
#https://www.googleapis.com/customsearch/v1/?key=AIzaSyCATL1zbY0i3KDaIKfwz6t6peMyT_vA-0E&cx=012773643409177149611:giuhxzmolrc

#Take user input
keyword = input("search a keyword ")

#Connect to google.com and search the keyword
google_url = "https://www.googleapis.com/customsearch/v1/?key=AIzaSyCATL1zbY0i3KDaIKfwz6t6peMyT_vA-0E&cx=012773643409177149611:giuhxzmolrc"
search_request = google_url + "&q=" +keyword

#Send the request to server
import requests
response = requests.get(search_request)

#Show the response from Google
print(response.text)
```

```
#Filter
#Convert text response from google api's into JSON
import json
json_response = json.loads(response.text)
json_response["context"]

#Filter searchTime and Total Results
json_response["searchInformation"]["searchTime"]

#Filter formatted and htmlSnippet
#From all the items in the json_response

from wordcloud import WordCloud
import matplotlib.pyplot as plt
import re

for item in json_response["items"]:
  url = item["formattedUrl"]
  html = item["htmlSnippet"]
  html = re.sub("<.*?>","",html)
  html = re.sub("&nbsp","",html)
  
  #Filter HTML into natural text
  wc = WordCloud().generate(html)
  plt.imshow(wc)
  plt.axis("off")
  plt.title(url)
  print(url)
  
  
  
  
  
```
